## Анализ данных о бронировании отелей

```python
import pandas as pd
```

1. Импортируйте библиотеку pandas как pd. Загрузите датасет bookings.csv с разделителем ";". Проверьте размер таблицы, типы переменных, а затем выведите первые 7 строк, чтобы посмотреть на данные. 

```python
df = pd.read_csv('bookings.csv', sep=';')
```

2. Приведите названия колонок к нижнему регистру и замените пробелы на знак нижнего подчеркивания.

```python
def my_rename(text):
    return text.lower().replace(" ", "_")
```

```python
_columns = {val: my_rename(val) for val in df.columns.tolist()}
```

```python
df = df.rename(columns=_columns)
```

```python
df.head()
```

3. Пользователи из каких стран совершили наибольшее число успешных бронирований? Укажите топ-5.

```python
df \
    .query('is_canceled == 0') \
    .groupby('country') \
    .size() \
    .sort_values(ascending=False) \
    .head(5)
```

4. На сколько ночей в среднем бронируют отели разных типов?

```python
df.groupby('hotel').agg({'stays_total_nights':'mean'})
```

5. Иногда тип номера, полученного клиентом (assigned_room_type), отличается от изначально забронированного (reserved_room_type). Такое может произойти, например, по причине овербукинга. Сколько подобных наблюдений встретилось в датасете?

```python
df.query('assigned_room_type != reserved_room_type').shape[0]
```

6.На какой месяц чаще всего успешно оформляли бронь в 2016? Изменился ли самый популярный месяц в 2017? <br>

```python
most_popular_month_2016 = df \
    .query('arrival_date_year == 2016 & is_canceled == 0') \
    .groupby('arrival_date_month') \
    .size().idxmax()
most_popular_month_2016
```

```python
most_popular_month_2017 = df \
    .query('arrival_date_year == 2017 & is_canceled == 0') \
    .groupby('arrival_date_month') \
    .size().idxmax()
most_popular_month_2017 == most_popular_month_2016
```

7. Сгруппируйте данные по годам и проверьте, на какой месяц бронирования отеля типа City Hotel отменялись чаще всего в каждый из периодов.

```python
grouped = df \
    .query('is_canceled == 1 & hotel == "City Hotel"') \
    .groupby(['arrival_date_year', 'arrival_date_month']) \
    .size().reset_index(name='count') \
    .sort_values(['arrival_date_year', 'count'], ascending=[True, False])
grouped
```

```python
maxes = grouped \
    .groupby(['arrival_date_year'], as_index=False) \
    .agg({'count' : 'max'})
keys = list(maxes.columns.values)
i1 = grouped.set_index(keys).index
i2 = maxes.set_index(keys).index
grouped[i1.isin(i2)]
```

8. Посмотрите на числовые характеристики трёх переменных: adults, children и babies. Какая из них имеет наибольшее среднее значение?

9. Создайте колонку total_kids, объединив children и babies. Отели какого типа в среднем пользуются большей популярностью у клиентов с детьми?

```python
df['total_kids'] = df['children'] + df['babies']
```
